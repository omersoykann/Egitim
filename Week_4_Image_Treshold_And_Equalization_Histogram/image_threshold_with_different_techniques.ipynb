```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
my_img_rgb = cv2.imread("./sau.jpg",1)
my_img_gray = cv2.cvtColor(my_img_rgb,cv2.COLOR_BGR2GRAY)
```

```python
_, my_thresh_binary = cv2.threshold(my_img_gray ,thresh=128, maxval=255,
                                    type=cv2.THRESH_BINARY)
```

```python
fig = plt.figure(figsize=(10,7))
rows = 1
colums = 2
fig.add_subplot(rows,colums,1)
plt.imshow(my_img_gray, cmap="gray")
plt.axis('off')
plt.title("original Image")

fig.add_subplot(rows,colums,2)
plt.imshow(my_thresh_binary, cmap="gray")
plt.axis('off')
plt.title("BINARY THRESHOLDING Image")
```

```python
_, my_thresh_binary = cv2.threshold(my_img_gray ,thresh=128, maxval=255,
                                    type=cv2.THRESH_BINARY_INV)
```

```python
fig = plt.figure(figsize=(10,7))
rows = 1
colums = 3
fig.add_subplot(rows,colums,1)
plt.imshow(my_img_gray, cmap="gray")
plt.axis('off')
plt.title("original Image")

fig.add_subplot(rows,colums,3)
plt.imshow(my_thresh_binary, cmap="gray")
plt.axis('off')
plt.title("BINARY INV THRESHOLDING Image")
```

```python
_, my_thresh_trunc = cv2.threshold(my_img_gray ,thresh=128, maxval=255,
                                    type=cv2.THRESH_TRUNC)
```

```python
fig = plt.figure(figsize=(10,7))
rows = 2
colums = 2
fig.add_subplot(rows,colums,1)
plt.imshow(my_img_gray, cmap="gray")
plt.axis('off')
plt.title("original Image")

fig.add_subplot(rows,colums,2)
plt.imshow(my_thresh_binary, cmap="gray")
plt.axis('off')
plt.title("BINARY  THRESHOLDING Image")

fig.add_subplot(rows,colums,3)
plt.imshow(my_thresh_binary, cmap="gray")
plt.axis('off')
plt.title("BINARY INV THRESHOLDING Image")

fig.add_subplot(rows,colums,4)
plt.imshow(my_thresh_binary, cmap="gray")
plt.axis('off')
plt.title("TRUNC THRESHOLDING Image")
```

```python
_, my_thresh_tozero = cv2.threshold(my_img_gray ,thresh=128, maxval=255,
                                    type=cv2.THRESH_TOZERO)
_, my_thresh_tozero = cv2.threshold(my_img_gray ,thresh=128, maxval=255,
                                    type=cv2.THRESH_TOZERO_INV)
```

```python
fig = plt.figure(figsize=(10,7))
rows = 2
colums = 3
fig.add_subplot(rows,colums,1)
plt.imshow(my_img_gray, cmap="gray")
plt.axis('off')
plt.title("original Image")

fig.add_subplot(rows,colums,2)
plt.imshow(my_thresh_binary, cmap="gray")
plt.axis('off')
plt.title("BINARY  THRESHOLDING Image")

fig.add_subplot(rows,colums,3)
plt.imshow(my_thresh_binary, cmap="gray")
plt.axis('off')
plt.title("BINARY INV THRESHOLDING Image")

fig.add_subplot(rows,colums,4)
plt.imshow(my_thresh_binary, cmap="gray")
plt.axis('off')
plt.title("TRUNC THRESHOLDING Image")

fig.add_subplot(rows,colums,5)
plt.imshow(my_thresh_binary, cmap="gray")
plt.axis('off')
plt.title("TOZERO THRESHOLDING Image")

fig.add_subplot(rows,colums,6)
plt.imshow(my_thresh_binary, cmap="gray")
plt.axis('off')
plt.title("TOZER INV THRESHOLDING Image")

plt.tight_layout()
plt.savefig("./my_all_threshold_techniques.png",dpi=300)
plt.show()
```
